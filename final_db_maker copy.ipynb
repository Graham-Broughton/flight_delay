## delay/plane

```python
flights['total_delay'] = flights['dep_delay'] + flights['arr_delay']
```

```python
delay = pd.DataFrame(flights['total_delay'].groupby(flights['op_unique_carrier']).mean()).sort_values('total_delay', ascending=False).reset_index()


# arrival = pd.DataFrame(flights['arr_delay'].groupby(flights['op_unique_carrier']).mean()).sort_values('arr_delay', ascending=False).reset_index()

# delay = delay.merge(arrival, how='left', on='op_unique_carrier')
```

```python
delay.describe()
```

```python
rating = {} #gonna rate these by q
for x,y in zip(delay['op_unique_carrier'], delay['total_delay']):
    if (y < 3):
        rating[x] = 0 #no delay
    elif (y >= 3) and (y < 12):
        rating[x] = 1 #low delay
    elif (y >= 12) and (y < 22):
        rating[x] = 2 #medium delay 
    elif (y >= 22):
        rating[x] = 3 #high, q3 of means
 
```

```python
airline_delay = pd.DataFrame([rating])
```

```python
final = airline_delay.T
```

```python
final
```

```python
final.to_csv("G_PSQL_data/airline_delay_rating.csv")
```
